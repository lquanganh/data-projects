```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read data

df_artist = pd.read_csv("artists.csv")

df_track = pd.read_csv("tracks.csv")
```

```python
#The most popularity song

df_track.loc[(df_track["popularity"]==df_track["popularity"].max()), ["name", "popularity", "artists", "release_date"]]
```

```python
#Top 10 songs popularity

df_track.nlargest(10, "popularity")[["name", "popularity", "artists", "release_date"]]
```

```python
#Change "release_date" type to "date_time"

df_track["release_date"].info()
```

```python
df_track["release_date"] = pd.to_datetime(df_track["release_date"])
```

```python
df_track["release_date"].info()
```

```python
#Check 18th position

df_track.iloc[18]
```

```python
#Change column "duration_ms" to column "duration"

df_track["duration"] = df_track["duration_ms"].apply(lambda x: round(x/1000))
df_track.drop("duration_ms", inplace=True, axis=1)
```

```python
df_track.head()
```

```python
#Correlation heatmap between variable

corr_df = df_track.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")

plt.figure(figsize=(14, 6))

heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, center=0, cmap="inferno",
                     linewidths=1, linecolor="Black");
heatmap.set_title("Correlation heatmap between variable");
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90);
```

```python
#Split data

sample_df = df_track.sample(int(0.004*len(df_track)))
```

```python
len(sample_df)
```

```python
#Regression plot

plt.figure(figsize=(14, 6))

sns.regplot(data=sample_df, y="loudness", x="energy", color="c").set(title="Loudness vs Energy")
```

```python
plt.figure(figsize=(14, 6))

sns.regplot(data=sample_df, y="popularity", x="acousticness", color="r").set(title="Popularity vs Acousticness")
```

```python
#Extract year from datetime object

df_track["year"] = df_track["release_date"].dt.year
```

```python
df_track.head()
```

```python
sns.displot(df_track["year"], discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr = df_track["duration"]

fig_dims = (18, 7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=df_track["year"], y=total_dr, ax=ax, errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90);
```

```python
sns.set_style(style="whitegrid")
fig_dims = (12, 5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x=df_track["year"], y=total_dr, ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60);
```

```python
num_df = df_track.select_dtypes(include="number")
```

```python
num_df
```

```python
plt.style.use("seaborn")

names = list(num_df.columns)

plot_per_row = 2

fig, ax = plt.subplots(round(len(names)/2), plot_per_row, figsize=(15, 25))

y=0;
for name in names:
    i, j = divmod(y, plot_per_row)
    sns.histplot(x=df_track[name], kde=True, ax=ax[i, j], color="purple")
    y = y+1

plt.tight_layout()
plt.show()
```

```python
plt.style.use("seaborn")

names = list(num_df.columns)

plot_per_row = 2

fig, ax = plt.subplots(round(len(names)/2), plot_per_row, figsize=(15, 25))

y=0;
for name in names:
    i, j = divmod(y, plot_per_row)
    sns.boxplot(x=df_track[name], ax=ax[i, j], palette="Set3")
    y = y+1
    
plt.tight_layout()
plt.show()
```
