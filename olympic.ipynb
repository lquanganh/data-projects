```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
athletes = pd.read_csv("athlete_events.csv")
athletes.head()
```

```python
region = pd.read_csv("noc_regions.csv")
region.head()
```

```python
athletes_df = athletes.merge(region, how = "left", on = "NOC")
athletes_df.head()
```

```python
athletes_df.shape
```

```python
#change columns name

athletes_df = athletes_df.rename(columns={"region":"Region",
                           "notes":"Notes"})
athletes_df.head()
```

```python
athletes_df.describe()
```

```python
#Top countries paticipating


ax = athletes_df["Team"].value_counts().head(10)
```

```python
#Plot top 10 countries

ax.plot(kind="bar", figsize=(14, 6))
```

```python
#Age distribution of athletes

plt.figure(figsize=(14, 6))
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Number of paticipants")
plt.hist(athletes_df["Age"], bins=np.arange(10, 80, 2), color="orange", edgecolor="white");
```

```python
athletes_df.head()
```

```python
#Female paticipants each olympics

female = athletes_df.loc[(athletes_df["Sex"]=="F") & (athletes_df["Season"]=="Summer"), ["Season", "Sex", "Year"]]
female["Year"].value_counts().sort_values(ascending=False)
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x="Year", data=female, palette="Spectral")
plt.title("Woman paticipants")
```

```python
part = female.groupby("Year")["Sex"].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:, "F"].plot()
plt.title("Plot of female athletes over time")
```

```python
#Gold Medal only


gold_medal = athletes_df[(athletes_df["Medal"]=="Gold")]
gold_medal.head()
```

```python
gold_medal[np.isfinite(gold_medal["Age"])].head()
```

```python
#Gold medal beyond 60


gold_medal[gold_medal["Age"] > 60]
```

```python
gm = gold_medal[gold_medal["Age"] > 60]["Sport"].value_counts()
```

```python
gm.plot(kind="bar", figsize=(14, 6))
```

```python
#Gold medal for each country


gold_medal["Region"].value_counts()
```

```python
totalgm = gold_medal["Region"].value_counts().reset_index(name="Medal").head(6)

g = sns.catplot(x="index", y="Medal", data=totalgm, height=5, kind="bar", palette="rocket")
g.despine(left=True)
g.set_xlabels("Top 6 country")
g.set_ylabels("Number of Medals")
plt.title("Medal for each country")
```
