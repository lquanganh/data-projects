```python
all_data = pd.read_csv("all_data.csv")
all_data.tail()
```

```python
all_data["Month"] = all_data["Order Date"].str[0:2]
```

```python
all_data.iloc[90000:90050]
```

```python
all_data.isnull()
```

```python
all_data.isnull().sum()
```

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df
```

```python
all_data = all_data.dropna()
```

```python
all_data.isnull().sum()
```

```python
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
```

```python
all_data["Month"] = all_data["Month"].astype("int32")
```

```python
all_data.tail()
```

```python
all_data.info()
```

```python
#converts columns into another types:
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"], downcast="integer")
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"], downcast="float")

all_data.info()
```

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
all_data.head()
```

```python
highest_month = all_data.groupby("Month").sum()
```

```python
month = range(1, 13)

plt.bar(month, highest_month["Sales"])
plt.xticks(month)
plt.ylabel("Sales in US ($)")
plt.xlabel("Month number")
plt.show()
```

```python
#split sentences:

def get_state(x):
    return x.split(",")[2].split(" ")[1]

all_data["City"] = all_data["Purchase Address"].apply(lambda x: f"{x.split(',')[1]} ({get_state(x)})")
all_data.head()
```

```python
hightest_city = all_data.groupby("City").sum()
```

```python
c = hightest_city["Sales"]
c.plot.bar(figsize=(10,5)).set_ylabel("Million")
```

```python
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
```

```python
all_data.head()
```

```python
all_data.info()
```

```python
all_data["Hour"] = all_data["Order Date"].dt.hour
all_data["Minute"] = all_data["Order Date"].dt.minute
```

```python
all_data.head()
```

```python
hours = [hour for hour, df in all_data.groupby("Hour")]

plt.plot(hours, all_data.groupby("Hour").count())
plt.xticks(hours)
plt.xlabel("Hour")
plt.ylabel("Number of Order")
plt.grid()
plt.show()
```

```python
df = all_data[all_data["Order ID"].duplicated(keep=False)]
df.head(15)
```

```python
df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))
df = df[["Order ID", "Grouped"]].drop_duplicates()
df.head(15)
```

```python
df.head(15)
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in df["Grouped"]:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list, 2))) #Number 2 can change, 2 means count 2 object go along with each others, 3 means count 3....

print(count)
```

```python
count.most_common(10)
```
